#Creating A function to do Normalization

```python
import re
def remove_punctuation_lower(text):
  text=re.sub('[^a-zA-Z0-9]',' ',text)
  text=text.lower()
  return text
```

```python
df['text']=df['text'].apply(remove_punctuation_lower)
df.head()
```

#REMOVING STOPWORD

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
sw=stopwords.words('english')
```

```python
nltk.download('punkt_tab')
```

```python
def remove_stop_words(text):
  words=[]
  wordlist=nltk.word_tokenize(text)
  for word in wordlist:
    if word not in sw:
      words.append(word)
  return ' '.join(words)
```

```python
df['text']=df['text'].apply(remove_stop_words)
```

```python
df
```

#Stemming

```python
from nltk.stem import PorterStemmer
stemmer =PorterStemmer()

def stemming(text):
  stemmed = []
  words = nltk.word_tokenize(text)
  for word in words:
    stemmed.append(stemmer.stem(word))
  return ' '.join(stemmed)
```

```python
df['text']=df['text'].apply(stemming)
df.head()
```

```python
df.iloc[44894]['text']
```

#Define dependent and independent variable

```python
x=df['text']
y=df['target']
```

#Splitting Training And Testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

#convert text to vector
like label encoding converting text to numeric

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
x_train = tf.fit_transform(x_train)# learn vocabulary and transform training data
x_test = tf.transform(x_test)
```

```python
x_train[:5]
```

#Naive Bayas Algo

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(x_train,y_train)
```

```python
nb.score(x_test,y_test)
```

#LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

#DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

#RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```
